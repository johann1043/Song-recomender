```python
from config import *
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd

c_id = client_id
c_se = client_secret

sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=c_id, client_secret=c_se))
```

## Get Playlist

```python

# get Playlist

def get_playlist_tracks(username, playlist_id):
    playlist = {}
    results = sp.user_playlist_tracks(username,playlist_id,market="GB")
    tracks = results['items']
    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])
    playlist = tracks
    get_ids(playlist)
    

# get all Playlist ids


def get_ids(playlist):
    bp_ids = []
    for i in playlist:
        bp_ids.append(i['track']['id'])
    create_chunks(bp_ids)


# chunck the Playlist ids in 100

def create_chunks(bp_ids):
    bp_ids_chopped = []
    chunk_size = 100
    bp_ids_chopped = [bp_ids[i:i+chunk_size] for i in range(0,len(bp_ids), 100)]
    create_dataframe(bp_ids_chopped)


# create combined List

final_list = []
def create_dataframe(bp_ids_chopped):
    
    for i in bp_ids_chopped:
        final_list.extend(sp.audio_features(i))

# create DataFrame

def get_df(final_list):
    return pd.DataFrame(final_list)



# drop duplicates

def drop_dup_reset(data_frame):
    rows_before = len(data_frame)
    data_frame.drop_duplicates(inplace=True)
    data_frame.reset_index(drop=True, inplace=True)
    rows_after = len(data_frame)
    num_dups_deleted = rows_before - rows_after
    print(f'Deleted {num_dups_deleted} duplicates')

    return data_frame
```
